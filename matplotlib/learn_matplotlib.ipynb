## plt.plot(展现变量的趋势变化)
```plt.plot(x,y,ls='-',lw=2,label='plot figure')```
- x x轴上的数值
- y y轴上的数值
- ls 折线图的线条风格
- lw 折线图的线条宽度
- label 标记图形内容的文本标签

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.05,10,10000)
y = np.cos(x)

plt.plot(x,y,ls='--',lw=3,label='plot figure')
plt.legend()
plt.show()
````

## plt.scatter(寻找变量之间的关系)
```plt.scatter(x,y,c="b",label="scatter figure")```
- x x轴上的数值
- y y轴上的数值
- c 散点图中的颜色
- label 标记图形内容的文本标签

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.05,10,1000)
y = np.random.rand(1000)
plt.scatter(x,y,c=y,label="scatter figure")
plt.legend()
plt.show()
````

## grid 绘制刻度线的网格线
```plt.grid(linestyle = ":", color = "r")```
- linestyle 网格线的线条风格
- color 网格线的线条颜色

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.05,10,1000)
y = np.exp(x)

plt.plot(x,y,ls="--",lw=2,c = "c",label ="plot figure")
plt.legend()

plt.grid(linestyle=":",color='r')
plt.show()
````

## axhline axvline 绘制平行于X Y 轴的水平参考线
```plt.axhline(y=0.0,c='r',ls='--',lw=2)```
- y 水平参考线的出发点
- c 水平参考线的颜色
- ls 水平参考线的风格
- lw 水平参考线的宽度

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.05,10,1000)
y = np.exp(x)

plt.plot(x,y,ls="--",lw=2,c = "c",label ="plot figure")


plt.grid(linestyle=":",color='r')

plt.axvline(5,c='g',ls=':',lw=3,label="axvline label")
plt.axhline(6000,c="r",ls="--",lw=3,label="axhine label")

plt.legend()
plt.show()
````

## axvspan axhspan 绘制垂直于X Y轴的参考区域
```plt.axvspan(xmin=1.0,xmax=2.0,facecolor='y',alpha='0.3')```
- xmin 参考区域的起始位置
- xmax 参考区域的终止位置
- facecolor 参考区域的填充颜色
- alpha 参考区域的填充颜色的透明度

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.05,10,1000)
y = np.exp(x)

plt.plot(x,y,ls="--",lw=2,c = "c",label ="plot figure")


plt.grid(linestyle=":",color='r')

plt.axvline(5,c='g',ls=':',lw=3,label="axvline label")
plt.axhline(6000,c="r",ls="--",lw=3,label="axhine label")

plt.axvspan(2,4,facecolor='y',alpha=0.3)
plt.axhspan(11111,22222,facecolor='g',alpha=0.5)
plt.annotate('local max', xy=(8, 2611), xytext=(+10, +30),textcoords='offset points', ha='right', va='bottom',bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))
plt.legend()
plt.show()
````

```plt.annotate('local max', xy=(x_local_max, y_local_max), xytext=(+10, +30),textcoords='offset points', ha='right', va='bottom',bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))```

## 函数组合应用

````python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm as cm

# 创建数据
x = np.linspace(0.5,3.5,100)
y = np.sin(x)
y1 = np.random.randn(100)

# 散点图
plt.scatter(x,y1,c="0.25",label="scatter figure")

# 线图
plt.plot(x,y,ls='-',lw=2,label="line figure")

# 其它 去除边框
for spine in plt.gca().spines.keys():
    if spine == "top" or spine == "right":
        plt.gca().spines[spine].set_color("none")
plt.gca().xaxis.set_ticks_position("bottom")
plt.gca().yaxis.set_ticks_position("left")

# 设置x,y轴的范围
plt.xlim(0,4)
plt.ylim(-3,3)

# 设置x,y轴的标签
plt.xlabel("x_axis")
plt.ylabel("y_axis")

# 设置网格
plt.grid(True,ls=":",color="0.25")

# 增加水平参考线
plt.axhline(y=0,c='r',ls='--',lw=2,label="y=0")

# 增加垂直参考线
plt.axvline(x=3.5,c='r',ls='--',lw=2,label="x=0")

# 增加垂直区域
plt.axvspan(1,2,facecolor="y",alpha=0.3,label="x=1~2")

# 设置指向文本
plt.annotate("maximum",xy=(np.pi/2,1),xytext=((np.pi/2)+0.15,1.5),weight="bold",color='r',arrowprops=dict(arrowstyle="->",connectionstyle="arc3",color='0.25'))
plt.annotate("spines",xy=(0.75,-3),xytext=(0.35,-2.25),weight="bold",color='b',va='bottom',ha='center',arrowprops=dict(arrowstyle="->",connectionstyle='arc3',color='b'))
plt.annotate("",xy=(0,-2.78),xytext=(0.4,-2.32),arrowprops=dict(arrowstyle="->",connectionstyle="arc3",color="b"))
plt.annotate("",xy=(3.5,-2.98),xytext=(3.6,-2.70),arrowprops=dict(arrowstyle="->",connectionstyle="arc3",color="b"))

# 设置文本
plt.text(3.6,-2.70,"'|' is tickline",weight="bold",color="b")
plt.text(3.6,-2.95,"3.5 is ticklabel",weight="bold",color="b")

# 设置标题
plt.title("Learn matplotlib")
plt.legend(loc="upper left")
# 显示图形
plt.show()
````

# 二、使用统计函数绘制简单图形

## bar 用于绘制柱状图
```plt.bar(x,y)```

参数说明：
- x：横坐标，列表或数组
- y：纵坐标，列表或数组

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl 

# mpl.rcParams["font.sans-serif"] = ["SimHei"]
# mpl.rcParams["axes.unicode_minus"] = False

x = [1,2,3,4,5,6,7,8]
y = [1,4,9,16,25,36,49,64]

plt.bar(x,y,align="center",color='c',tick_label=['a','b','c','d','e','f','g','h'],hatch='/')
# plt.bar(x,y,align="center",color='c',hatch='/')

plt.xlabel("code")
plt.ylabel("Weight(kg)")
plt.title("Bar Chart")

plt.show()
````

## barh 用于绘制横向条形图
```plt.barh(x,y)```

参数说明：
- x: 标示在y轴上的数据的类别
- y: 每种数据的类别的数量

````python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4,5,6,7,8]
y = list(np.random.randint(2,10,8))

plt.barh(x,y,align='center',color='r',tick_label=['a','b','c','d','e','f','g','h'],hatch='/')
plt.xlabel('Y')
plt.ylabel('X')
plt.title('Bar Chart')

plt.show()
````

## hist 用于绘制直方图
plt.hist(x)

参数说明：

- x：在X轴上绘制箱体的定量数据输入值
- bins：直方图的分段数目，默认为10
- range：数据范围，默认为数据范围
- normed：是否归一化，默认为False
- weights：权重，默认为None
- cumulative：是否计算累计频数，默认为False
- bottom：直方图的下边界，默认为0
- histtype：直方图的类型，默认为直方图

````python
import matplotlib.pyplot as plt
import numpy as np

box_weight = np.random.randint(0,100,100)
x = box_weight.astype(int)

bins = range(0,11,1)

plt.hist(x,bins=bins,color='g',histtype="bar",rwidth='1',alpha=0.6)

plt.xlabel('Box Weight')
plt.ylabel('Frequency')
plt.title('Box Weight Distribution')
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)  # 生成随机数据

plt.figure(figsize=(10, 4))

plt.subplot(141)
plt.hist(data, bins=30, color='b', alpha=0.6, histtype='bar')
plt.title('Bar Histogram')

plt.subplot(142)
plt.hist(data, bins=30, color='g', alpha=0.6, histtype='barstacked')
plt.title('Bar Stacked Histogram')

plt.subplot(143)
plt.hist(data, bins=30, color='r', alpha=0.6, histtype='step')
plt.title('Step Histogram')

plt.subplot(144)
plt.hist(data, bins=30, color='y', alpha=0.6, histtype='stepfilled')
plt.title('Step Filled Histogram')

plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

data1 = np.random.randn(1000)  # 第一个数据序列
data2 = np.random.randn(1000)  # 第二个数据序列

plt.figure(figsize=(10, 4))

plt.subplot(121)
plt.hist([data1, data2], bins=30, color=['b', 'g'], alpha=0.6, label=['Data 1', 'Data 2'])
plt.legend()
plt.title('Bar Histogram')

plt.subplot(122)
plt.hist([data1, data2], bins=30, color=['b', 'g'], alpha=0.6, histtype='barstacked', label=['Data 1', 'Data 2'])
plt.legend()
plt.title('Bar Stacked Histogram')

plt.tight_layout()
plt.show()
````

## pie 用于绘制饼状图

## polar 用于绘制极线图
```plt.polar(theta,r)```

参数说明：

- theta：极角，单位为弧度
- r：极半径

````python
import numpy as np
import matplotlib.pyplot as plt

bar_num = 12

theta = np.linspace(0,2*np.pi,bar_num,endpoint=False)
r = 30*np.random.rand(bar_num)

# 使用极坐标方式绘制图形
plt.polar(theta,r, color="chartreuse", linewidth=2, marker="*", mfc='b', ms=10)

plt.show
````

## scatter 用于绘制散点图
```plt.scatter(x, y, s=100, c='r', marker='o', alpha=0.5)```
参数说明：
- x，y：用于绘制散点的 x，y 坐标。
- s：散点大小，以像素为单位。
- c：散点颜色。
- marker：散点形状，可以是单个字符，也可以是字符串，例如 “*” “.” “x” “+” “,” “|” “_” “^” “<” “>” “v” “D” “H” “d” “
- cmap：颜色映射，用于指定颜色映射的名称。
- alpha：透明度，取值范围为 0 到 1。

````python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

x = np.random.randn(100)
y = np.random.randn(100)

plt.scatter(x,y,s=np.power(10*x+20*y,2),c=np.random.rand(100),cmap=mpl.cm.RdYlBu,marker='o')

plt.show()
````

## errorbar 用于绘制误差棒图
```plt.errorbar(x, y, yerr=None, xerr=None, fmt='none', ecolor='gray', elinewidth=1, capsize=1, barsabove=False, lolims=False, uplims=False, data=None, **kwargs)```

参数说明：

- x：横坐标
- y：纵坐标
- yerr：纵坐标误差
- xerr：横坐标误差
- fmt：图形标记
- ecolor：误差线颜色
- elinewidth：误差线宽度
- capsize：误差线尾巴大小
- barsabove：是否在误差线上方绘制误差线

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.1,0.6,6)
y = np.exp(x)

plt.errorbar(x,y,fmt='bo:',yerr=0.2,xerr=0.02)
plt.xlim(0,0.7)
plt.show()
````
